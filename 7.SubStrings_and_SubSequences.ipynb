# SubString and Sequences

### Return the substrings of a string

```python
def substrings(s, i, j, result):
    if i == len(s):   # base case: done with all start positions
        return

    if j > len(s):    # if end passes string, move to next start
        substrings(s, i+1, i+1, result)
        return

    result.append(s[i:j])   # store substring
    substrings(s, i, j+1, result)  # expand end pointer

    
    
s = "abc"
result = []
substrings(s, 0, 1, result)

print(result)
```

```python
def substrings(string, start, end, result):
    
    if start == len(string):
        return
    
    if end == len(string) + 1:
        substrings(string, start+1, start+1, result)
        return     # <-- VERY IMPORTANT
    
    result.append(string[start:end])
    
    substrings(string, start, end+1, result)
    

s = 'abc'
result = []

substrings(s, 0, 1, result)
print(result)
```

### Return the subsequences of a string

```python
def subsequence(string):
    
    if string == '':
        return ['']
    
    smallAns = subsequence(string[1:])
    
    myChar = string[0]
    
    ans = []
    ans.extend(smallAns)
        
    for i in smallAns:
        ans.append(myChar + i)
        
    return ans

subsequence('abc')
```
