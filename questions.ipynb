```python
def generate_square(n):
    a=[]
    """
    Function to return a square pattern of '*' of side n as a list of strings.
    
    Parameters:
    n (int): The size of the square.
    
    Returns:
    list: A list of strings where each string represents a row of the square.
    """
    for i in range(n):
        a.append('*' *n)
    return a

print(generate_square(3))
print(generate_square(5))
```

```python
# function to count the occurrences of each item in a list:

def item_count(lst):
    b = {}
    for item in lst:
        if item not in b:
            b[item] = 1
        else:
            b[item] =b[item] + 1
    return b

item_count([1,1,2,3,5,5,5,2])
```

```python
# Reverse string in a list
def word_reverse(String):
    rev_words = []
    
    words = String.split()
    for word in words:
        rev_words.append(word[::-1])
        
    sentence = ' '.join(rev_words)
    return sentence

String = 'This is Abi' 

print(word_reverse(String))
```

```python
# to find if a number is prime or not

def prime_number(num):
    
    n = int(num **0.5) # considering the sq root 
    # if there is no divisible number with its square root number it will be a prime number
    
    if n <2:
        return False
    
    for i in range(2,n+1):
        if num%i ==0:
            return False
    return True

num = 5
res = prime_number(num)
if res is True:
    print(f"{num} is prime number ")
else:
    print(f"{num} is not a prime number")
```

```python
# Flatten a nested list

def flatten(lst):
    
    flat_list = []
    
    for i in lst:
        
        if isinstance(i,list): # if item in a list is of type list it will run agin the flatten function
            flat_list.extend(flatten(i))
        
        else:
            flat_list.append(i)
    
    return flat_list


lst = [3, 4, [1, 3, [8, 0, 1], 7], 6, [12, 55]]
print(flatten(lst))
```

```python
# Sum of an array

def Sum(arr):
    
    total = 0
    n = len(arr)
    
    if n == 0:
        return 0
    
    else:
        
        for i in range(n):
            total = total + arr[i]
            
    return total

arr = [1,2,3,4,5]
arr2 = []

print(Sum(arr))
print(Sum(arr2))
```

```python
def recursive_sum(arr):
    """
    Function to calculate the sum of an array using recursion.
    
    Parameters:
    arr (list of int): The array of integers.
    
    Returns:
    int: The sum of the array elements.
    """
    if len(arr) == 0:
        return 0
        
    
    
    return arr[0] + recursive_sum(arr[1:])


recursive_sum([1,2,3,4,5])
    
    
```

```python
def find_first_index(arr, element):
    """
    Function to find the first index of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    int: The first index of the element in the array, or -1 if not found.
    """
    
    
    if len(arr) == 0:
        return -1
        
    if arr[0] == element:
        return 0
        
    ans = find_first_index(arr[1:], element) 
    
    if ans == -1:
        return ans
        
    else:
        return ans +1
        
    return ans


print(find_first_index([1, 2, 3, 2, 4, 2], 2))
```

```python
# find the last index of an element

def last_index(arr, x):
    for i in range(len(arr)-1, -1, -1): # range start, end, step -> starts from len(arr) ends in 0 backwards
        if arr[i] == x:
            return i
    return -1   # if element not found

print(last_index([1, 2, 3, 2, 5], 2))
```

```python
def find_last_index(arr, element,index= 0):
    """
    Function to find the last index of a given element in an array using recursion.
    
    Parameters:
    arr (list of int): The array to search through.
    element (int): The element to find.
    
    Returns:
    int: The last index of the element in the array, or -1 if not found.
    """
    
    
    if len(arr) == index:
        return -1
        

        
    ans = find_last_index(arr, element,index+1  ) 
    
    if ans != -1:
        return ans
    
    if arr[index] == element :
        return index

        
    return ans
    
print(find_last_index([1, 6, 3,3], 3))
```

```python
def merge_sort(arr):
    """
    Function to perform merge sort on a list of integers using recursion.
    
    Parameters:
    arr (list of int): The list to be sorted.
    
    Returns:
    list of int: The sorted list.
    """
    # Your code here
    
    if len(arr) <= 1:
        return arr 
        
    middle = len(arr) //2
    
    left = merge_sort(arr[:middle])
    right = merge_sort(arr[middle:])
    
    return merge(arr, left, right)
    

def merge(arr, left, right):
    
    
    result = []
    
    i = 0
    j = 0
    
    while (i < len(left) )and ( j < len(right)):
        
        if left[i] < right[j]:
            result.append(left[i])
            
            i+= 1 
            
        else:
            result.append(right[j])
            
            j+= 1
            
    result.extend(left[i:])
    result.extend(right[j:])
    
    return result
    
arr = [5, 2, 9, 1, 5, 6]    
merge_sort(arr)
```
