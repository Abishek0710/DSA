# Binary Tree

## Printing a binarytree

```python
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def printTree(root):

    if root is None:
        return
    
    print(f"Root: {root.data}")
    

    # Format : Node : L->LeftChildData , R->RightChild Data
    if root.left is not None:
        print(f"L-> {root.left.data} ", end=', ')
    else:
        print("L-> None ", end=', ')

    if root.right is not None:
        print(f"R-> {root.right.data} ", end=' ')
    else:
        print("R-> None ", end=' ')

```

```python
root = BinaryTreeNode(1)
root.left = BinaryTreeNode(2)
root.right = BinaryTreeNode(3)

printTree(root)
```

## Take input level wise for a binary tree

```python
from collections import deque

def takeinput_levelwise():

    data = int(input('Enter data for the node : '))

    if data == -1:
        return None
    
    root = BinaryTreeNode(data)

    q = deque([root])


    while len(q) != 0:

        currentnode = q.popleft()

        leftchilddata = int(input(f"Enter the left child data for {currentnode.data}: "))
        if leftchilddata != -1:
            leftnode = BinaryTreeNode(leftchilddata)
            currentnode.left = leftnode
            q.append(leftnode)

        rightchilddata = int(input(f"Enter the right child data for {currentnode.data}: "))
        if rightchilddata != -1:
            rightnode = BinaryTreeNode(rightchilddata)
            currentnode.right = rightnode
            q.append(rightnode)

    return root


root = takeinput_levelwise()
printTree(root)
```

## Cound nodes in a Binary Tree

```python
def countnodes(root):

    if root is None:
        return 0
    
    leftcount = countnodes(root.left)
    rightcount = countnodes(root.right)

    return leftcount + rightcount + 1
```

## Diameter of a Tree

```python

# Height of a tree
def height(root):

    if root is None:
        return 0
    
    leftheight = height(root.left)
    rightheight = height(root.right)

    return max(leftheight, rightheight) + 1 # Adding 1 to add the current node



def diameter(root):

    if root is None:
        return 0
    
    leftheight = height(root.left)
    rightheight = height(root.right)
    
    leftdiameter = diameter(root.left)
    rightdiameter = diameter(root.right)



    return max(leftheight + rightheight, leftdiameter, rightdiameter) # Diameter can be either in left subtree, right subtree or passing through root
```
