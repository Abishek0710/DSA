# Tree Data Structure

```python
from collections import deque



class Tree:
    def __init__(self, data):
        self.data = data
        self.children = []

# We have created a structure for a tree node with data and children attributes.

# We have to define if the node is a root node or a child node.


# Here, we have created a root node with data 10 and three child nodes with data 20, 30, and 40 respectively.
# The child nodes are added to the root node's children list.
# This forms a simple tree structure.


def print_tree(root):
    if root is None:
        return
    print(root.data)

    for eachchild in root.children:
        print_tree(eachchild)



def print_tree_optimized(root):
    if root is None:
        return
    print(f'{root.data}: ', end = "")

    for eachchild in root.children:
        print(f'{eachchild.data}', end = ",") 

    print()

    for eachchild in root.children:
        print_tree_optimized(eachchild)


# Take input ans assign for nodes and children

def take_input():

    data = int(input("Enter data for the node: "))

    root = Tree(data)

    queue = deque([root])

    while len(queue) != 0:

        currentNode = queue.popleft()

        num_children = int(input(f"enter the no of child node for {currentNode.data}: "))

        for i in range(num_children):
            child_data = int(input(f"Enter the data for child {i+1} of node {currentNode.data}: "))

            childNode = Tree(child_data)

            currentNode.children.append(childNode)

            queue.append(childNode)

    return root




root = take_input()
print_tree_optimized(root)
```
