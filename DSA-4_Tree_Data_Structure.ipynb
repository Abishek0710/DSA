# Tree Data Structure

```python
class Tree:
    def __init__(self, data):
        self.data = data
        self.children = []

# We have created a structure for a tree node with data and children attributes.

# We have to define if the node is a root node or a child node.


# Here, we have created a root node with data 10 and three child nodes with data 20, 30, and 40 respectively.
# The child nodes are added to the root node's children list.
# This forms a simple tree structure.


def print_tree(root):
    if root is None:
        return
    print(root.data)

    for eachchild in root.children:
        print_tree(eachchild)



def print_tree_optimized(root):
    if root is None:
        return
    print(f'{root.data}: ', end = "")

    for eachchild in root.children:
        print(f'{eachchild.data}', end = ",")

    print()

    for eachchild in root.children:
        print_tree_optimized(eachchild)



root = Tree(1)

child1 = Tree(2)
child2 = Tree(3)
child3 = Tree(4)

root.children.append(child1) # root->c1
child1.children.append(child2) # c1->c2
root.children.append(child3) # root->c3

# root --> c1
#      --> c3
# c1 --> c2
# c2 --> None
# c3 --> None
print_tree_optimized(root)
```
