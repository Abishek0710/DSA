# Recursion in Strings

### Reverse a string using recursion

```python
def reverse(string):
    
    if len(string) == 0:
        return string
    
    small_ans = string[1:]
    
    return small_ans + string[0]

reverse('Abishek')
```

### Remove all the occurances of a character in a string

```python
def removeChar(string,char):
    
    if len(string) == 0 or string == '':
        return string
    
    small_ans = removeChar(string[1:],char)
    
    if string[0] == char:
        return small_ans
    else:
        return string[0] + small_ans
    
    return small_ans

removeChar('Abizz','z')
```

### Check if a string is palindrome or not using recursion

```python
def CheckPalindrome(string1):
    
    string = string1.lower()
    
    return PalindromeHelper(string, start= 0, end = len(string) - 1)

def PalindromeHelper(string, start, end):
    
    if start> end:
        return True
    
    if string[start] != string[end]:
        return False
    
    return PalindromeHelper(string, start+1, end-1)
    
print(CheckPalindrome('Nitin') )  
print(CheckPalindrome('malayalam'))
```
