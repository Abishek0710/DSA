# Recursion

### Head Recursion

```python
def headRecursion(num):
    
    if num == 0:
        return
    
    headRecursion(num-1)
    
    print(num)
    
headRecursion(5)
```

### Tail Recursion

```python
def tailRecursion(num):
    
    print(num)
    
    if num == 0:
        return
    tailRecursion(num-1)
    
tailRecursion(5)
```

### Find the Factorial of a number using Recursion

```python
def Factorial(n):
    
    if n == 1:
        return 1
    
    return n * Factorial(n-1)

Factorial(5)
```

### Find the Fibonacci series of a number

```python
def Fibonacci(n):
    
    if n == 0:
        return 0
    
    if n == 1:
        return 1
    
    return Fibonacci(n-1) + Fibonacci(n-2)

Fibonacci(10)
```
