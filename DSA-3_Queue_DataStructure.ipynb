# Queue Data Structure

It follows the principle of "First in, First out" (FIFO), where the first element added to the queue is the first one to be removed.

It is used as a buffer in computer systems where we have speed mismatch between two devices that communicate with each other. For example, CPU and keyboard and two devices in a network

Queue is also used in Operating System algorithms like CPU Scheduling and Memory Management, and many standard algorithms like Breadth First Search of Graph, Level Order Traversal of a Tree.

#### Queue contains basic operations like enqueue, dequeue, size, front, isEmpty

enqueue - To add data to Queue

size - Displays no of data present in the Queue

isEmpty - Returns True if Queue is empty else False

Front/peek - returns the first inserted data in the Queue

dequeue - deletes the first data of the stack and displays the deleted element

```python
class QueueUsingList():
    
    def __init__(self):
        self.__queue = []
        
    def enqueue(self,data):
        self.__queue.append(data)
        return f'Added {data} to the Queue'
    
    def dequeue(self):
        if self.size() == 0:
            return 'Queue is Empty,Cannot Dequeue'
        
        return self.__queue.pop(0)
    
    def size(self):
        return len(self.__queue)
    
    def front(self):
        if self.size() == 0:
            return 'Queue is Empty,Cannot show Front'
        return self.__queue[0]
    
    def isEmpty(self):
        return self.size() == 0
    
Q = QueueUsingList()

Q.enqueue(10)
Q.enqueue(20)
Q.enqueue(30)
print(Q.size())
print(Q.isEmpty())
print(Q.front())
print(Q.dequeue())
print(Q.dequeue())
print(Q.front())
print(Q.size())
```
