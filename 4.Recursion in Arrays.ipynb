### Write a program to check if an array is sorted or not

```python
def checkSorted(arr):
    n = len(arr)
    
    if n == 0 or n == 1:
        return True
    
    if arr[0] > arr[1]:
        return False
    
    return checkSorted(arr[1:])
    

arr = [1,2,3,4,5,6,6,7,8] 

checkSorted(arr)
```

```python
### Sum of elements of a array using recursion

def sumOfArray(arr):
    
    n = len(arr)
    
    if n == 0:
        return 0
    
    ans = arr[0] +  sumOfArray(arr[1:])
    return ans

arr = [1,2,3,4,5]

sumOfArray(arr)
```

### Given an array and an element find the first index of the element using recursion

```python
def firstIndex(arr, ele):
    
    if len(arr) == 0:
        return -1
    
    if arr[0] == ele:
        return 0
    
    ansfromrecursion = firstIndex(arr[1:], ele)

    
    if ansfromrecursion == -1:
        return ansfromrecursion
    
    else:
        return ansfromrecursion + 1
    

arr = [1,2,3,4,5]

firstIndex(arr, 3)
```

### Given an array and an element find the first index of the element using recursion without slicing array

```python
def firstIndex(arr, ele, index = 0):

    
    if index == len(arr):
        return -1
    

    if arr[index] == ele:
        
        return index
    
    
    ans = firstIndex(arr, ele, index+1)
    
    return ans


arr = [1,2,4,3,5,3]
firstIndex(arr, 3)
    
```

### Given an array and an element find the last index of the element using recursion without slicing array
