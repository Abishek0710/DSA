#### In array memory is contiguous so memory address of elements will be in increasing order of 4 bytes(32 bits)

#### The memory address formula for an element is = Base memory + (position of element  x  32)

#### ie. if base memory is 10000 the for 3rd element = 10000 + (3 x 32) = 10096

#### While using a linked list memory address is non-contiguous wherever spare is found elements are stared there

```python
class Node:
    
    def __init__(self,value):
        self.data = value
        self.next = None

first = Node(1)
second = Node(5)
third = Node(6)

first.next = second
second.next = third

print(first.next.data) # gives the value of second node
print(second.next.data) # gives value of third node

print(first.next.next.data) # gives the value of third node
```

# Creating a Linked list which takes values from user

```python
class Node:
    def __init__(self,value):
        self.data = value
        self.next = None

def take_input_better():
    value = int(input("Enter the value of Node :- "))
    head = None
    tail = None

    while(value != -1):
        newNode = Node(value)
        if(head == None):
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode

        value = int(input("Enter the value of Node :- "))

    return head

def print_LL(head):
    temp = head
    while(temp!=None):
        print(temp.data,end="->")
        temp = temp.next

    print()
    return

newhead = take_input_better()
print_LL(newhead)
```

```python
class node():
    
    def __init__(self,value): # initializing a constuructor
        self.data = value
        self.next = None
        
def take_input():
    value = int(input("Enter input: ")) # should be int else when giving -1 as input it will consider as string leading to infinite loop
    head = None
    tail = None
    
    while (value != -1):
        newnode = node(value)
        
        if head == None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode
        value = int(input("Enter input: "))


    return head

def print_LL(head):
    
    temp = head
    while temp != None:
        
        print(temp.data,end = '->')
        temp = temp.next
        
newhead = take_input()
print_LL(newhead)
```

## Length of a linked List

```python
# function for length of linked list
# the head node returned in Linked List algorithm above is used aas input here

def length_Of_LL(head):
    
    temp = head
    count = 0
    while temp != None:
        count = count +1
        print(f'{temp.data}, {count}')
        temp = temp.next
        

length_Of_LL(newhead)
```

## Length of list in recursive way

```python
def Length_LL_recursive(head):
    
    if head == None:
        return 0
    
    smallans = Length_LL_recursive(head.next)
    
    return 1 + smallans

Length_LL_recursive(newhead)
```
