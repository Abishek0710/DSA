```python
class Node:
    
    def __init__(self,value):
        self.data = value
        self.next = None

first = Node(1)
second = Node(5)
third = Node(6)

first.next = second
second.next = third

print(first.next.data) # gives the value of second node
print(second.next.data) # gives value of third node

print(first.next.next.data) # gives the value of third node
```

# Creating a Linked list which takes values from user

```python
class node():
    
    def __init__(self,value): # initializing a constuructor
        self.data = value
        self.next = None
        
def take_input():
    value = int(input("Enter input: ")) # should be int else when giving -1 as input it will consider as string leading to infinite loop
    head = None
    tail = None
    
    while (value != -1):
        newnode = node(value)
        
        if head == None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode
        value = int(input("Enter input: "))


    return head

def print_LL(head):
    
    temp = head
    while temp != None:
        
        print(temp.data,end = '->')
        temp = temp.next
    
    print('None')
        
newhead = take_input()
print_LL(newhead)
```

## Length of a linked List

```python
# function for length of linked list
# the head node returned in Linked List algorithm above is used aas input here

def length_Of_LL(head):
    
    temp = head
    count = 0
    while temp != None:
        count = count +1
        print(f'{temp.data}, {count}')
        temp = temp.next
        

length_Of_LL(newhead)
```

## Length of list in recursive way

```python
def Length_LL_recursive(head):
    
    if head == None:
        return 0
    
    smallans = Length_LL_recursive(head.next)
    
    return 1 + smallans

Length_LL_recursive(newhead)
```

## Insert at the head of a linked list

```python
## Insert at the head of a linked list

def insert_at_head_LL(head, value):
    
    newNode = node(value)
    newNode.next = head
    head = newNode
    
    return head


    

new = take_input()
print_LL(new)

newhead = insert_at_head_LL(new,100)
print('--------After inserting at head--------')
print_LL(newhead)
```

```python
## Insert at the head of a linked list
## Taking insert value from user

def insert_at_head_LL_better(head):
    
    value = int(input('Enter a value for inserting at Head Node : '))
    
    if value == -1:
        return head
    else:
        newNode = node(value)
        newNode.next = head
        head = newNode
    
    return head


    

new = take_input()
print_LL(new)

newhead = insert_at_head_LL_better(new)
print('--------After inserting at head--------')
print_LL(newhead)
```

## Insert at tail of Linked list

```python
def insert_at_tail_LL(head):
    
    value = int(input('Enter value for tail node : '))
    newNode = node(value)
    
    if head is None:
        return newNode
    
    temp = head
    while ( temp.next != None):
        temp = temp.next
    
    temp.next = newNode
    
    return head

new_head = insert_at_tail_LL(new)
print_LL(newhead)
```

## Insert at tail recursively

```python
def insert_at_tail_recursively(head, value):
    
    if head is None:
        return node(value)
    
    head.next = insert_at_tail_recursively(head.next, value)
    return head

value = int(input("Enter value for tail node: "))
new_head = insert_at_tail_recursively(newhead, value)
print_LL(new_head)
```

## Insert at any index

```python
def insert_at_index(head, value, index):
    
    newNode = node(value)
    
    if index == 0:
        newNode.next = head
        head = newNode
        return head
    
    temp = head
    count = 0
    
    while temp is not None and count < index -1:
        temp = temp.next
        count = count+1
    
    if temp is None:
        print('Index out of Bounds //n Cannot proceed with insert operation')
        return head
    
    newNode.next = temp.next
    temp.next = newNode
    
    return head

print_LL(newhead)
newhead = insert_at_index(newhead, 100, 2)
print('------------------After Inserting------------------')
print_LL(newhead)
        
```

## Insert at any Index - Recursively

```python
def insert_at_index_recursively(head, value, index):
    
    

    if index == 0:
        newNode = node(value)
        newNode.next = head
        head = newNode        
        return head
    
    if head == None:
        print('Index out of bounds')
        return head
    
    head.next = insert_at_index_recursively(head.next, value, index-1)
    
    return head


print_LL(newhead)
newhead = insert_at_index_recursively(newhead, 200, 2)
print('------------------After Inserting------------------')
print_LL(newhead)
    
```

# Deleting a node

## Deleting a node at head

```python
def del_head_node(head):
    if head is None:
        return None
    newhead = head.next
    return newhead

print_LL(newhead)
newhead = del_head_node(newhead)
print('----------After Deleting---------')
print_LL(newhead)
```

## Deleting a Tail Node

```python
def del_tail_node(head):
    
    if head is None or head.next is None:
        return None # If there is one Node there is no tail Node
    
    temp = head
    
    while temp.next.next is not None:
        temp = temp.next
        
    print(f'Stopped at node for deleting, {temp.data}')
    temp.next = None
    return head


print_LL(newhead)
newhead = del_tail_node(newhead)
print('----------After Deleting---------')
print_LL(newhead)
```

## Delete a tail Node - recursively

```python
def del_tail_node_recursively(head):
    
    if head is None:
        return None
    if head.next is None:
        return None
    
    head.next = del_tail_node_recursively(head.next)
    
    return head

print_LL(newhead)
newhead = del_tail_node_recursively(newhead)
print('----------After Deleting---------')
print_LL(newhead)
```

## Delete a node of any index

```python
def del_node_by_index(head,index):
    
    if head == None:
        return None
    
    if index == 0:
        return head.next
    
    temp = head
    count = 0
    
    while temp is not None and count < index-1:
        temp = temp.next
        count = count +1
    
    if temp is None or temp.next is None:
        print('Out of Bounds')
        return head
    

    print(temp.next.data)
    print(temp.next.next.data)
    
    temp.next = temp.next.next
    return head

print_LL(newhead)
newhead = del_node_by_index(newhead,10)
print('----------After Deleting---------')
print_LL(newhead)
```

## delete a node by index recursively 

```python
def del_node_by_index_recursively(head, index):

    # index out of bounds
    if head is None:
        print("Index Out of Bounds")
        return head

    # delete at index 0
    if index == 0:
        return head.next

    head.next = del_node_by_index_recursively(head.next, index - 1)
    return head

print_LL(newhead)
newhead = del_node_by_index_recursively(newhead,5)
print('----------After Deleting---------')
print_LL(newhead)
```

## Delete node by value

```python
def del_node_by_val(head, value):
    
    if value == -1:
        return
    if head.data == value:
        return head.next
    
    temp = head
    
    while temp.next is not None and temp.next.data != value:
        temp= temp.next
        
    if temp.next is None:
        print('Value not present')
        return head
    
    temp.next = temp.next.next
    return head


print_LL(newhead)
newhead = del_node_by_val(newhead,50)
print('----------After Deleting---------')
print_LL(newhead)

        
```

## Delete by value recursive

```python
def del_by_val_recursive(head, value):
    
    if head.data == value:
        return head.next
    
    if head.next == None:
        print('Out of bounds')
        return head
    
    head.next = del_by_val_recursive(head.next, value)
    
    return head

print_LL(newhead)
newhead = del_by_val_recursive(newhead,40)
print('----------After Deleting---------')
print_LL(newhead)
```

## Search a Node by value 

```python
def search_node_by_val(head, value):
    
    temp = head
    index = 0
    
    while temp != None:
        
        if temp.data == value:
            return index
        
        temp = temp.next
        index = index+1
        
    return 'Node not found'


print_LL(newhead)
print('----------Searching---------')
print(search_node_by_val(newhead,30))
```

## Search a node by index

```python
def search_node_by_val(head, index):
    
    temp = head
    count = 0
    
    while temp is not None and count != index:
        temp = temp.next
        count = count+1
        
    if temp == None or temp.next ==None:
        return 'Node Not Found'
    
    if count == index:
        return temp.data
    
print_LL(newhead)
print('----------Searching---------')
print(search_node_by_val(newhead,0))
```
