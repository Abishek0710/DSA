# Linear Search

```python
def LinearSearch(arr, target):
    
    for i in range(len(arr)):
        
        if i == target:
            
            return i
    return 'Not Found'

print(LinearSearch([3,4,5,6,7,3,1,1,2], 6))
print(LinearSearch([3,4,5,6,7,3,1,1,2], 9))
```

# Binary Search

Bineary Search work only on sorted array

```python
def BinarySearch(arr, target):
    
    start = 0
    end = len(arr) -1
    
    while start <= end:
        
        mid = (start + end ) //2
        
        if target < arr[mid]:
            end = mid -1
        if target > arr[mid]:
            start = mid  +1
            
        if target == arr[mid]:
            return mid
    return 'Not Found'

print(BinarySearch([1,2,3,4,5,6], 5))

print(BinarySearch([1,2,3,4,5,6], 9))
```
