# Sorting Algorithms

## Bubble Sort

```python
def BubbleSort(arr):
    
    n = len(arr) 
    
    for passes in range(n):
        
        for i in range(0, n-1-passes):
            
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
                
    return arr
                
arr = [3,4,2,4,6,7]

BubbleSort(arr)
```

## Selection Sort

```python
def SelectionSort(arr):
    
    n = len(arr)
    
    for i in range(n):
        
        current_index = i
        
        for j in range(i+1, n):
            
            if arr[j] < arr[current_index]:
                
                current_index = j
                
        arr[i], arr[current_index] = arr[current_index], arr[i]
                
    return arr
                

arr = [3,4,2,4,6,7]

SelectionSort(arr)
```

## Insertion Sort

```python
def insertionSort(arr):
    
    n = len(arr)
    
    for i in range(1, n):
        
        temp = arr[i]
        j = i - 1
        
        # Shift all elements greater than temp to the right
        while j >= 0 and arr[j] > temp:
            
            arr[j+1] = arr[j]
            j = j - 1
        
        # Place the value temp at its correct position
        arr[j+1] = temp
    
    return arr


arr = [3, 4, 2, 4, 6, 7]
print(insertionSort(arr))
```
